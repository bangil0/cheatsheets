## Displaying Images

```python
from skimage.io import imread
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Display images from skimage

```python
from skimage import data
coins = data.coins()
imgplot = plt.imshow(coins, cmap='hot')
plt.colorbar() # plot a color bar


# cmap can be ['Set1', 'Set2', 'Set3', 'hot', 'nipy_spectral']
```

#### Display external images

https://matplotlib.org/users/image_tutorial.html

```python
egimage = 'http://uforest.org/images/artocarpus_elasticus1.jpg'
img = imread(egimage) # render image as numpy array
plt.imshow(img)
plt.show()
```

#### Data Descriptions

```python
# data type
print(type(img))
# geometry
print(img.shape)
# pixel count
print(img.size)
# data type
print(img.dtype)
```
